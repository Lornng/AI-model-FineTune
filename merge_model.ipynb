# Load the adapter model and merge with base model


```python
import torch
from peft import AutoPeftModelForCausalLM
adapter_model = "Lornng/Llama2-7B-QLoRA-cpgQA"

trained_model = AutoPeftModelForCausalLM.from_pretrained(
    adapter_model,
    low_cpu_mem_usage=True,
    return_dict=True,
    torch_dtype=torch.float16,
    device_map={"": 0},
)
```

```python
# Merge LoRA with the base model and save the merged model
lora_merged_model = trained_model.merge_and_unload()
lora_merged_model.save_pretrained("Llama2-7b-merged-qlora-200-cpgQA",safe_serialization=True)

#push merged model to the hub
lora_merged_model.push_to_hub("Llama2-7b-merged-qlora-200-cpgQA")
```

```python
from transformer import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('meta-llama/Llama-2-7b')
tokenizer.push_to_hub('Llama2-7b-merged-qlora-200-cpgQA')
```
