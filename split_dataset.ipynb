```python
from datasets import load_dataset


# Split the data into 20% test, 80% training

data = load_dataset("Lornng/cpgQA-textcol", split='train')
data = data.train_test_split(test_size=0.2)
```

```python
data['train'].to_csv('cpgQA-textcol_train.csv')
```

```python
test_val = data['test']
test_val = test_val.train_test_split(test_size=0.5)
```

```python
test_val['test'].to_csv('cpgQA-textcol_test.csv')
test_val['train'].to_csv('cpgQA-textcol_validation.csv')
```
